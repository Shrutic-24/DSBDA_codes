```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
features=df.columns
features
```

```python
feature_types=df.dtypes
feature_types
```

```python
for feature in features:
    print(feature)
    plt.figure(figsize=(8,6))
    plt.hist(df[feature].values, alpha=0.7)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title('Histogram of'+feature)
    plt.show()
```

```python
data1 = df.drop('Id', axis=1)
data1.boxplot(figsize=(10, 10), patch_artist=True)
```
